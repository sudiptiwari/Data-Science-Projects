### mpg and kmpl are similar metrics. Converting kmpl of data_json to mpg using formula: 
### mpg = kmpl * 2.35215

```python
# Function to convert kmpl to mpg
def convert_to_mpg(kmpl):
    return kmpl * 2.35215

data_json['mpg'] = data_json['kmpl'].apply(convert_to_mpg)

# Round the 'mpg' column to 2 decimal places
data_json['mpg'] = data_json['mpg'].round(2)
```

```python
# Remove column 'kmpl'
data_json = data_json.drop(columns=['kmpl'])

data_json
```

### Rename Column name 'hp' of data_json to 'horsepower'

```python
# Renaming the column
# inplace = True makes changes into dataframe and updates the dataframe too
data_json.rename(columns={'hp': 'horsepower'}, inplace=True) 
```

```python
data_json
```

### Rename Column name 'country' of data_json to 'origin'

```python
data_json.rename(columns={'country': 'origin'}, inplace = True)
```

```python
data_json
```

### Rename Column name 'Name' of data_json to 'name'

```python
data_json.rename(columns={'Name': 'name'}, inplace = True)
```

```python
data_json
```

### Convert 'ground_clearance' to 'ground_clearance_mm'

```python
def convert_cm_to_mm(cm):
    return cm * 10

data_json['ground_clearance_mm'] = data_json['ground_clearance'].apply(convert_cm_to_mm)
```

```python
# Remove old column 'ground_clearance'
data_json.drop(columns = ['ground_clearance'])

data_json
```

### Convert displacement to float data type

```python
# Convert 'displacement' column to float
data_json['displacement'] = data_json['displacement'].astype(float)
data_json
```

# Feature Selection and Engineering

#### Features that have a high absolute correlation with the target variable are generally considered more important, while those with low correlation might be candidates for exclusion.

#### - Keep Features with High Absolute Correlation:
#### Keep features with high absolute correlation (e.g., |correlation| > 0.5).


#### - Skip Features with Low Absolute Correlation:

#### Skip features with low absolute correlation (e.g., |correlation| < 0.5).



```python
corr_matrix
```

```python
target_corr = corr_matrix['mpg'].sort_values(ascending=False)
print(target_corr)
```

### Here, we select the features whose values is greater than 0.5 in magnitude

- abs(target_corr):

This computes the absolute value of each correlation coefficient in target_corr. Absolute values are used to consider both positive and negative correlations.



- abs(target_corr) > threshold:

This creates a boolean Series where each value is True if the absolute correlation is greater than the specified threshold and False otherwise. This helps in filtering out correlations that are below the threshold.



- target_corr[abs(target_corr) > threshold]:

This applies the boolean mask to target_corr, returning only those correlations that have an absolute value greater than the threshold.



- .index:

This retrieves the index (i.e., the feature names) of the filtered Series. These are the names of the features that have a high absolute correlation with the target variable.



- .tolist():

This converts the index object into a list of feature names.

```python
# Define a threshold for high correlation
threshold = 0.5

# Select features with high absolute correlation with the target variable
important_features = target_corr[abs(target_corr) > threshold].index.tolist()
print("Important features:", important_features)

# Drop the target variable from the list of important features
important_features.remove('mpg')

# Select and print the features to be kept
print("Selected features for modeling:", important_features)
```

# Feature Normalization

### Feature normalization is a preprocessing step used in machine learning to adjust the scales of the input features, ensuring that they contribute equally to the learning process. This is particularly important for algorithms that are sensitive to the scale of the input data, such as gradient descent-based methods (e.g., linear regression, logistic regression, neural networks) and distance-based methods (e.g., k-nearest neighbors, support vector machines).



### StandardScaler (Standardization):
Standardization transforms the data to have a mean of 0 and a standard deviation of 1. This is also known as Z-score normalization.

Formula: 

x' = (xâˆ’Î¼)/Ïƒ
 
where, 

- Î¼ is the mean and 

- Ïƒ is the standard deviation.

### MinMaxScaler (Normalization):
Min-max normalization transforms the data to fit within a specified range, typically [0, 1].

Formula:

x'= (xâˆ’min(x))/(max(x)âˆ’min(x))

```python
rng = np.random.default_rng(seed = 1234)
x_unnormalized = rng.integers(low = 0, high = 100, size = 50)
x_unnormalized
```

### StandardScalar (Standardization)

```python
x_mean = np.mean(x_unnormalized)
x_std = np.std(x_unnormalized)
x_standardScalar = (x_unnormalized - x_mean)/x_std
x_standardScalar
```

### MinMaxScalar (Normalization)

```python
x_max = np.max(x_unnormalized)
x_min = np.min(x_unnormalized)
x_minMaxScalar = (x_unnormalized - x_min) / (x_max - x_min)
x_minMaxScalar
```

### One Hot Encoder

#### OneHotEncoder is a method in machine learning used to convert categorical variables into a format that can be provided to ML algorithms to do a better job in prediction. The idea is to encode categorical integer features as one-hot vectors. This is useful because many machine learning algorithms cannot work with categorical data directly and require all input variables to be numerical.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data = {
    "Rec-no" : range(10),
    "Temperature" : ["Hot", "Cold", "Very Hot",
                    "Warm", "Hot", "Warm",
                    "warm", "hot", "hot", "cold"],
    "Color": ["Red", "Yellow", "Blue",
             "Blue", "Red", "Yellow",
             "Red", "Yellow", "Yellow", "Yellow"],
    "Target" : [1,1,1,0,1,0,1,0,1,1]
}
df = pd.DataFrame(data)
df
```

### Case of Temperature varies. Hence we have to convert all values to lower case

```python
df['Temperature'] = df['Temperature'].str.lower()
df
```

```python
ohe = OneHotEncoder(sparse_output = False)
data_ohe = ohe.fit_transform(df[['Temperature']])
data_ohe
```

#### The OneHotEncoder in Scikit-Learn determines the order of the columns based on the sorted order of the unique values in the input categorical feature. 

```python
# ohe_df = pd.DataFrame(data_ohe, columns = ['temp_cold','temp_hot','temp_veryhot','temp_warm'])
# ohe_df

# Get feature names directly from OneHotEncoder
feature_names = ohe.get_feature_names_out(['Temperature'])

# Create a DataFrame with the encoded data
ohe_df = pd.DataFrame(data_ohe, columns=feature_names)

# Reset the index to align with the original DataFrame
ohe_df.index = df.index

ohe_df
```

#### In the context of pd.concat([df, ohe_df], axis=1), the axis parameter specifies the direction in which to concatenate the DataFrames.

- axis=0: Concatenate along rows (i.e., one DataFrame below the other).

- axis=1: Concatenate along columns (i.e., one DataFrame next to the other).

```python
# Concat two dataframes
df_ohe_encoded_final = pd.concat([df, ohe_df], axis = 1)

# Drop 'Temperature' column
df_ohe_encoded_final.drop(columns = 'Temperature', axis = 1, inplace = True)

df_ohe_encoded_final
```

### Label Encoding

Label encoding is a technique used to convert categorical data into numerical labels. Each unique category value is assigned a unique integer based on alphabetical order or frequency. This method is straightforward and can be useful in various scenarios, particularly when the categorical variable is ordinal (i.e., the categories have a meaningful order).

- **When to Use Label Encoding**
1. Ordinal Categorical Data:

When the categorical data has an intrinsic order (e.g., 'low', 'medium', 'high').
Example: Education levels ('high school', 'bachelor's', 'master's', 'PhD').
Tree-Based Algorithms:

2. When using algorithms like decision trees, random forests, and gradient boosting, which can handle label encoded data directly.


3. Simple Models:

When the model can handle integer inputs and the categorical variable has a large number of categories (e.g., product IDs).



- **When to Avoid Label Encoding**
1. Nominal Categorical Data:

When the categorical data has no intrinsic order (e.g., colors, cities).
In such cases, one-hot encoding is typically preferred.


2. Distance-Based Algorithms:

### Ordinal Value Encoding

Ordinal encoding is a type of encoding used for categorical data where the categories have a meaningful order or ranking. Unlike one-hot encoding, which treats all categories as having no intrinsic order, ordinal encoding assigns a unique integer value to each category, preserving the order of the categories.

- **When to Use Ordinal Encoding**
1. Ordered Categories: When the categorical data has an inherent order. For example:

2. Education levels: high school < bachelor's < master's < PhD

3. Temperature levels: cold < warm < hot < very hot

4. Rating levels: poor < average < good < excellent

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Creating original Dataset
data = {
    "Rec-no" : range(10),
    "Temperature" : ["Hot", "Cold", "Very Hot",
                    "Warm", "Hot", "Warm",
                    "warm", "hot", "hot", "cold"],
    "Color": ["Red", "Yellow", "Blue",
             "Blue", "Red", "Yellow",
             "Red", "Yellow", "Yellow", "Yellow"],
    "Target" : [1,1,1,0,1,0,1,0,1,1]
}
df = pd.DataFrame(data)
df
```

```python
# Create Uniform value of Ordinal Variable
df['Temperature'] = df['Temperature'].str.lower()
df
```

```python
# Define Custom Categories Order
categories = [['cold', 'warm', 'hot', 'very hot']] # cold = 0, warm = 1, hot = 2, very hot = 3

ordinal_encoder = OrdinalEncoder(categories = categories)

df_ordinal_encoded = pd.DataFrame( data = ordinal_encoder.fit_transform(df[['Temperature']]) , columns = ['Encoded_Temperature'] )
```

### Frequency Encoding

Frequency encoding is a technique used to encode categorical variables based on the frequency (or count) of each category in the dataset. This method replaces each category with the number of times it appears in the dataset. Frequency encoding can be particularly useful when dealing with high cardinality categorical variables, as it retains the information about the distribution of categories.

- **When to Use Frequency Encoding**

1. High Cardinality Categorical Data: When categorical variables have many unique categories.
2. Tree-Based Models: Frequency encoding can be useful for tree-based algorithms (e.g., decision trees, random forests, gradient boosting) as these models can benefit from frequency information.

```python
data = {
    "Rec-no" : range(10),
    "Temperature" : ["Hot", "Cold", "Very Hot",
                    "Warm", "Hot", "Warm",
                    "warm", "hot", "hot", "cold"],
    "Color": ["Red", "Yellow", "Blue",
             "Blue", "Red", "Yellow",
             "Red", "Yellow", "Yellow", "Blue"],
    "Target" : [1,1,1,0,1,0,1,0,1,1]
}
df = pd.DataFrame(data)
df
```

```python
df['Temperature'] = df['Temperature'].str.lower()
df
```

```python
# Compute the frequency of each category in the 'color' column
color_freq = df['Color'].value_counts().to_dict()
print("\nFrequency of each category in 'color':")
print(color_freq)

# Map the frequencies to the 'color' column
df['color_encoded'] = df['Color'].map(color_freq)

print("\nFrequency Encoded DataFrame:")
print(df)
```

### - **fit_transform**
### The fit_transform method is a convenience method that combines the functionality of fit and transform into a single step. This method is used during the training phase of a model to learn the parameters from the training data and immediately apply the transformation.

### - **fit + transform**
### The fit method is used to learn the parameters from the data (e.g., mean and standard deviation for standard scaling), and the transform method applies the learned parameters to the data to transform it. This approach is useful when you need to apply the same transformation to both the training and testing datasets.

### - **transform**
### The transform method is used to apply the learned parameters (from fit) to the data. This method is typically used for preprocessing steps such as scaling, encoding, or dimensionality reduction.

### - **predict**
### The predict method is used in the context of machine learning models to generate predictions based on the input features. This method is called on a trained model to make predictions on new data.
