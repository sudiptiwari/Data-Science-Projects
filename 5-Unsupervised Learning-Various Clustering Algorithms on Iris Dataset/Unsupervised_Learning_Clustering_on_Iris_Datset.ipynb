```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
```

```python
X1, label1 = make_blobs(n_samples = 200, n_features = 2, centers = 2, cluster_std = 5, random_state = 123)
```

```python
X2, label2 = make_moons(n_samples = 200, noise = 0.08, random_state = 123)
```

```python
# plt.scatter(X1[:,0], X1[:,1], c=label1, s=50, cmap='viridis')
plt.scatter(X1[:,0], X1[:,1], s=50)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Synthetic Data Generated by make_blobs')
plt.show()
```

```python
# plt.scatter(X2[:,0], X2[:,1], c=label2, s=50, cmap='viridis')
plt.scatter(X2[:,0], X2[:,1], s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Synthetic Data Generated by make_moons')
plt.show()
```

### Here, in first case, it is unclear to identify suitable number of clusters but in second case, two number of clusters are best.

# **Data X1 (Blobs)**

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
kmeans = KMeans(n_clusters=2, random_state=123)
kmeans_labels = kmeans.fit_predict(X1)

# Visualize K-means clustering results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X1[:, 0], X1[:, 1], c=kmeans_labels, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
agglomerative = AgglomerativeClustering(n_clusters=2)
agglomerative_labels = agglomerative.fit_predict(X1)

# Visualize Agglomerative clustering results
plt.subplot(1, 2, 2)
plt.scatter(X1[:, 0], X1[:, 1], c=agglomerative_labels, s=50, cmap='viridis')
plt.title('Agglomerative Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Compute the linkage matrix
Z = linkage(X1, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z, truncate_mode='level', p=5)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample index or (Cluster size)')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
# Perform DBSCAN clustering
dbscan = DBSCAN(eps=2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X1)
```

```python
# Visualize DBSCAN clustering results
plt.figure(figsize=(10, 5))
plt.scatter(X1[:, 0], X1[:, 1], c=dbscan_labels, s=50, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# **Apply clustering in Iris Dataset**

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

## Plot the true labels

```python
# Plot the true labels
plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, s=50, cmap='viridis')
plt.title('Actual Classification')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Visualize K-means clustering results
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, s=50, cmap='viridis')
plt.title('K-means Clustering')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
```

```python
agglomerative = AgglomerativeClustering(n_clusters=3)
agglomerative_labels = agglomerative.fit_predict(X_scaled)

# Visualize Agglomerative clustering results
plt.subplot(1, 3, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=agglomerative_labels, s=50, cmap='viridis')
plt.title('Agglomerative Clustering')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Visualize DBSCAN clustering results
plt.subplot(1, 3, 3)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=dbscan_labels, s=50, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')

plt.tight_layout()
plt.show()
```

### Aggolometric clustering provided the cluster that is closest to the original distribution.
